```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 100  # Number of IBIs to generate
mean_ibi = 60      # Mean inter-beat interval in seconds (heart rate of 60 BPM)
std_ibi = 1        # Standard deviation of IBIs (minimal HRV)

# Generate IBIs
ibis = np.random.normal(loc=mean_ibi, scale=std_ibi, size=num_samples)

# Calculate heart rate from IBIs
heart_rate = 60 / ibis

# Plot IBIs and heart rate
plt.figure(figsize=(10, 10))

plt.subplot(3, 1, 1)
plt.plot(ibis, marker='o', linestyle='-')
plt.xlabel('Sample')
plt.ylabel('Inter-Beat Interval (seconds)')
plt.title('Inter-Beat Intervals')

plt.subplot(3, 1, 2)
plt.plot(heart_rate, marker='o', linestyle='-')
plt.xlabel('Sample')
plt.ylabel('Heart Rate (BPM)')
plt.title('Heart Rate')


inputSignal = np.loadtxt('noise.csv', delimiter=',', dtype=float)

plt.subplot(3, 1, 3)
plt.plot(inputSignal[25:300], marker='o', linestyle='-')
plt.xlabel('Sample')
plt.ylabel('Heart Rate (BPM)')
plt.title('Heart Rate')



plt.tight_layout()
plt.show()
```
